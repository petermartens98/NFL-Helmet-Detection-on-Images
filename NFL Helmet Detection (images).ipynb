```python
# Import neccesary modules

import numpy as np
import pandas as pd
import seaborn as sns
import cv2
import imageio
import subprocess
import os
from PIL import Image
from IPython.display import Video, display


%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
plt.rcParams['figure.dpi'] = 150

import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in label data for images
img_labels = pd.read_csv("nfl-impact-detection-files\\image_labels.csv")
img_labels.head()
```

```python
# Set name of our working image
img_name = img_labels['image'][0]
img_name
```

```python
# Define path to selected image
img_path = f"nfl-impact-detection-files//images//{img_name}"
```

```python
# Read in and plot the image
img = imageio.imread(img_path) 
plt.imshow(img)
plt.show()
```

```python
# Function to add boxes around player helmets
def add_img_boxes(image_name, image_labels):
    HELMET_COLOR = (0,0,0) # Black
    boxes = img_labels.loc[img_labels['image'] == img_name]
    for j, box in boxes.iterrows():
        color = HELMET_COLOR
        cv2.rectangle(img, (box.left, box.top), (box.left + box.width, box.top + box.height),
                      color, thickness=2)
    plt.imshow(img)
    plt.show()
```

```python
add_img_boxes(img_name, img_labels)
```
